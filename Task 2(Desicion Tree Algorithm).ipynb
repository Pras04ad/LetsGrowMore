## Step4 : Data Preparation

```python
iris = iris.drop('Id', axis=1)
iris.head()
```

```python
x = iris.iloc[:, 0:4]
x.head()
```

```python
y = (iris.iloc[:, 4])
y.head().to_frame()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
std = StandardScaler()
x = std.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.3, random_state = 1)
```

## Step5 : Model Creation

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
model.fit(x_train, y_train)
```

## Prediction using the Created Model

```python
y_pred = model.predict(x_test)
y_pred
```

## Step6 : Model Evaluation

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
acc = accuracy_score(y_test, y_pred)
print("The Accuracy of the Decision Tree Algorithms is : ", str(acc*100) + "%")
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
lst = iris['Species'].unique().tolist()
df_cm = pd.DataFrame(data = cm, index = lst, columns = lst)
df_cm
```
